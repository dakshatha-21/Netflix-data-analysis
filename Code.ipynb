```python
#Clean show_id and cast, keep first actor only
df['show_id']=df['show_id'].str.replace('^s','',regex=True)
df
```

```python
#converting date_added to datetime
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df

                                
```

```python
#converting release_year to integer
df['release_year']=df['release_year'].astype("int64",errors='ignore')
df
```

```python
# Create new column for minutes and seasons
df['duration_minutes'] = df['duration'].where(df['type'] == 'Movie').str.extract('(\d+)').astype('int64',errors="ignore")
df['season_count'] = df['duration'].where(df['type'] == 'TV Show').str.extract('(\d+)').astype('int64',errors="ignore")
```

```python
df
```

```python
df['duration_minutes']=df['duration_minutes'].fillna(0)
df['season_count']=df['season_count'].fillna(0)
```
